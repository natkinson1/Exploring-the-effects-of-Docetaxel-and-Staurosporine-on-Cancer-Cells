## Shape Descriptors

Now let's write some functions to create new shape descriptor features. Some shape descriptors include 

- Compactness
- roundness
- orientation
- convexity


For some of these function, I need some of the original variables from the data set. These are:
- Pixel position X
- Pixel position Y 

These have already been added into the data.

```python
def roundness(df):
    """Finds the roundess for the cell at each frame. The closer to 0 this value is,
    the more round it is. Returns variance of X-axis pixels and Y-axis pixels respectively."""
    
    var = []
    for i in range(df.shape[0]):
        
        total_x = 0
        n = int(df["num_pixels"][i])*2
        
        for k in range(0, n, 2):
            
            x = np.array([df["Pixel Position X (Pixels)"][i],
                          df["Pixel Position Y (Pixels)"][i]]) - np.array([df.iloc[i,k],
                                                                           df.iloc[i,k+1]])
            total_x += np.linalg.norm(x)
        
        var.append(total_x/int(df["num_pixels"][i]))
    
    return var

pixel_data["Var"] = roundness(pixel_data)
```

These two new columns var_x, var_y are telling me how spherical the cells are at each frame. I am calculating the variance between each x axis pixel with the overall pixel position X. The same is being done for the y positions. I finding the difference in distances for the perimeter of the cell compared with the centre of the cell, then finding the variance of all of these distances summed together.
```python
#Will need this feature for the compactness function
pixel_data["cell_area"] = data["Area (µm²)"]
```

```python
import math
def compactness(df):
    """Finds the area of the smallest box which encircles the cell.
    The compactness is a ratio of the area of the cell compared with the box."""
    
    #Find the max distance between for the boundary of the cell.
    data_frame = []
    box_area = []
    for i in range(df.shape[0]):
        max_dist = 0
        n = int(df["num_pixels"][i]*2)
        for j in range(0,n,2):
            for k in range(0,n-j,2):
                x_dist = (df.iloc[i,j]-df.iloc[i,k])**2
                y_dist = (df.iloc[i,j+1]-df.iloc[i,k+1])**2
                dist = np.sqrt(x_dist + y_dist)
                if dist > max_dist:
                    max_dist = dist
                    x_0 = df.iloc[i,j]
                    x_1 = df.iloc[i,k]
                    y_0 = df.iloc[i,j+1]
                    y_1 = df.iloc[i,k+1]
                else:
                    continue

        min_y = 500
        max_y = 0
        n = int(df["num_pixels"][i]*2)-1
        alpha = math.atan2(y_1 - y_0, x_1 - x_0)
        
        for k in range(0, n, 2):
            y = y_0 - math.sin(alpha)*(df.iloc[i,k] - x_0) 
            + math.cos(alpha)*(df.iloc[i,k+1]- y_0)
            if y < min_y:
                min_y = y
            elif y > max_y:
                max_y = y
            else:
                continue
        box_area.append(abs((x_1 - x_0)*(max_y-min_y)))
    
    return box_area/df["cell_area"]

pixel_data["compactness"] = compactness(pixel_data)
```

This function firstly finds a minimum box around the cell, this box is horizontal:

Then in the second part of the function, the box is rotated to find the true minimum box around the cell.

```python
def perimeter_length(df):
    """Finds the length of the perimeter for each cell."""
    perim_dis = []
    for i in range(df.shape[0]):
        total_len = 0
        n = int(df["num_pixels"][i]-1)
        for k in range(0,n,2):
            x_dis = (df.iloc[i,k]-df.iloc[i,k+2])**2
            y_dis = (df.iloc[i,k+1]- df.iloc[i,k+3])**2
            total_len += np.sqrt(x_dis + y_dis)
        perim_dis.append(total_len)
    return perim_dis

pixel_data["cell_perimeter"] = perimeter_length(pixel_data)
```
